```python
# Script para formatar geradores
#Script auto el√©trica souza LucasAmorimPorciuncula

def formatar_geradores(entrada):
    linhas = entrada.strip().split("\n")  
    resultado = []
    for linha in linhas:
        partes = linha.split(".") 
        if len(partes) > 1:
            num = partes[1].strip()  
            if num.isdigit():  
                # Determinar o KVA e o c√≥digo
                if len(num) == 4: 
                    kva = num[:2]  
                    codigo = f"GG{num[2:]}" 
                elif len(num) > 4:  
                    kva = num[:3]  
                    codigo = f"GG{num[3:]}"  
                else:
                    continue  
                resultado.append(f"{kva}KVA {codigo}")  
    return "\n".join(resultado)

# Input do usu√°rio
entrada = """
1. 11004
2. 5532
3. 5522
4. 5526
5. 50002
6. 50001
7. 50003
8. 50004
9. 27501
10. 18003
11. 11003
14. 5503
15. 5515
16. 17003
17. 26001
18. 5533
19. 17004
20. 5509
"""

# Processar a entrada
saida = formatar_geradores(entrada)

# Imprimir o resultado
print(saida)
```

# Script para Formatar Relat√≥rio no Estilo WhatsApp  

Este script permite que voc√™ preencha informa√ß√µes sobre geradores e formate o relat√≥rio no estilo de mensagens do WhatsApp.  

## Nome do Arquivo  
`formatador_relatorio_whatsapp.py`  

## Exemplo de Uso  
1. Execute o script.  
2. Preencha as informa√ß√µes solicitadas (nome do gerador, destino, data, motorista, etc.).  
3. Digite `sair` para finalizar.  
4. O relat√≥rio ser√° exibido no formato desejado.  

```python
from datetime import datetime

def saudacao():
    hora_atual = datetime.now().hour
    if hora_atual < 12:
        return "‚òÄÔ∏è Bom dia"
    elif hora_atual < 18:
        return "üå§Ô∏è Boa tarde"
    else:
        return "üåÉ Boa noite"

def formatar_relatorio():
    relatorio = []
    print(f"Preencha as informa√ß√µes dos geradores. Para finalizar, digite 'sair'.\n")
    
    while True:
        nome = input("Nome do gerador (ou 'sair' para finalizar): ")
        if nome.lower() == '':
            break
        destino = input(f"Destino de {nome}: ")
        data = input(f"Data de {nome} (dd/mm/aaaa): ")
        motorista = input(f"Motorista de {nome}: ")
        status = input(f"Status de {nome}: ")
        observacao = input(f"Observa√ß√£o de {nome} (pressione Enter se n√£o houver): ")
        regime_trabalho = input(f"Qual o regime de trabalho? 8h, 12h, 24h? ")

        gerador = {
            "nome": nome,
            "destino": destino,
            "data": data,
            "motorista": motorista,
            "status": status,
            "observacao": observacao,
            "regime_trabalho": regime_trabalho
        }
        relatorio.append(gerador)
    print(f'{saudacao()}, Dona Socorro! ')
    print("\n*Relat√≥rio Formatado:*\n")
    for gerador in relatorio:
        print(f"- *{gerador['nome']}*")
        print(f"  üìç Destino: {gerador['destino']}")
        print(f"  üìÖ Data: {gerador['data']}")
        print(f"  üöö Motorista: {gerador['motorista']}")
        print(f"  üîÑ Status: {gerador['status']}")
        print(f"  üïù Regime de trabalho: {gerador['regime_trabalho']}")
        if gerador['observacao']:
            print(f"  ‚ö†Ô∏è Observa√ß√£o: {gerador['observacao']}")
        print()  # Linha em branco entre os geradores

# Executa o script
if __name__ == "__main__":
    formatar_relatorio()
```

```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime
import urllib
import os
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# N√∫mero fixo do WhatsApp
NUMERO_FIXO = "71999740292" 

mensagens = []

def saudacao():
    hora = datetime.now().hour
    if hora < 12:
        return "‚òÄÔ∏è Bom dia"
    elif hora < 18:
        return "üå§Ô∏è Boa tarde"
    else:
        return "üåÉ Boa noite"

def adicionar_gerador():
    nome = entry_nome.get()
    destino = entry_destino.get()
    data = entry_data.get()
    motorista = entry_motorista.get()
    status = entry_status.get()
    observacao = entry_observacao.get()
    regime = entry_regime.get()

    if not nome:
        messagebox.showerror("Erro", "O campo 'Nome' √© obrigat√≥rio.")
        return

    msg = f"{saudacao()}, Dona Socorro!\n\n*Relat√≥rio do gerador:* *{nome}*\n" \
          f"üìç *Destino:* {destino}\nüìÖ *Data:* {data}\nüöö *Motorista:* {motorista}\n" \
          f"üîÑ *Status:* {status}\nüïù *Regime:* {regime}"
    if observacao:
        msg += f"\n‚ö†Ô∏è *Observa√ß√£o:* {observacao}"

    mensagens.append(msg)
    messagebox.showinfo("Adicionado", f"Gerador '{nome}' adicionado ao relat√≥rio.")

    # Limpar campos
    entry_nome.delete(0, tk.END)
    entry_destino.delete(0, tk.END)
    entry_data.delete(0, tk.END)
    entry_motorista.delete(0, tk.END)
    entry_status.delete(0, tk.END)
    entry_observacao.delete(0, tk.END)
    entry_regime.delete(0, tk.END)

def enviar_mensagens():
    if not mensagens:
        messagebox.showwarning("Aviso", "Nenhum gerador adicionado.")
        return

    try:
        servico = Service(ChromeDriverManager().install())
        options = webdriver.ChromeOptions()
        options.add_argument(r"user-data-dir=C:\Users\Lucas\AppData\Local\Temp\Profile Selenium")
        navegador = webdriver.Chrome(service=servico, options=options)

        navegador.get("https://web.whatsapp.com")
        messagebox.showinfo("Login", "Fa√ßa login no WhatsApp Web se ainda n√£o estiver logado.")
        while len(navegador.find_elements(By.ID, 'side')) < 1:
            sleep(1)
        sleep(2)

        for mensagem in mensagens:
            texto = urllib.parse.quote(mensagem)
            link = f"https://web.whatsapp.com/send?phone={NUMERO_FIXO}&text={texto}"

            navegador.get(link)
            while len(navegador.find_elements(By.ID, 'side')) < 1:
                sleep(1)
            sleep(2)

            try:
                btn = navegador.find_element(By.XPATH, '//*[@id="main"]/footer/div[1]/div/span/div/div[2]/div/div[4]/button/span')
                btn.click()
                sleep(5)
            except:
                messagebox.showerror("Erro", "Falha ao clicar no bot√£o de envio. Verifique se o n√∫mero √© v√°lido ou se est√° desbloqueado.")
                navegador.quit()
                return

        navegador.quit()
        messagebox.showinfo("Sucesso", "Todas as mensagens foram enviadas com sucesso.")
        mensagens.clear()

    except Exception as e:
        messagebox.showerror("Erro", str(e))

# Interface Gr√°fica com Tkinter
janela = tk.Tk()
janela.title("Relat√≥rio de Geradores")
janela.geometry("400x450")

# Campos
tk.Label(janela, text="Nome do Gerador*").pack()
entry_nome = tk.Entry(janela, width=50)
entry_nome.pack()

tk.Label(janela, text="Destino").pack()
entry_destino = tk.Entry(janela, width=50)
entry_destino.pack()

tk.Label(janela, text="Data (dd/mm/aaaa)").pack()
entry_data = tk.Entry(janela, width=50)
entry_data.pack()

tk.Label(janela, text="Motorista").pack()
entry_motorista = tk.Entry(janela, width=50)
entry_motorista.pack()

tk.Label(janela, text="Status").pack()
entry_status = tk.Entry(janela, width=50)
entry_status.pack()

tk.Label(janela, text="Observa√ß√£o").pack()
entry_observacao = tk.Entry(janela, width=50)
entry_observacao.pack()

tk.Label(janela, text="Regime de Trabalho").pack()
entry_regime = tk.Entry(janela, width=50)
entry_regime.pack()

tk.Button(janela, text="Adicionar Gerador", command=adicionar_gerador, bg="lightblue").pack(pady=10)
tk.Button(janela, text="Enviar pelo WhatsApp", command=enviar_mensagens, bg="lightgreen").pack(pady=10)

janela.mainloop()
```
